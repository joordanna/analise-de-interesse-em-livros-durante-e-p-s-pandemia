```python
import pandas as pd

df_livros_geral = pd.read_csv('livros_geral.csv',header = 1)

df_livros_geral.columns = ['mes', 'quantidade']

df_livros_geral['mes'] = pd.to_datetime(df_livros_geral['mes'], format='%Y-%m')
df_livros_geral['quantidade'] = pd.to_numeric(df_livros_geral['quantidade'])

df_livros_geral['tipo'] = 'geral'

df_livros_geral['mes_str'] = df_livros_geral['mes'].dt.strftime('%Y-%m')

print(df_livros_geral)
```

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

max_val = df_livros_geral['quantidade'].max()
min_val = df_livros_geral['quantidade'].min()

colors = []
for qtd in df_livros_geral['quantidade']:
    if qtd == max_val:
        colors.append('green')  
    elif qtd == min_val:
        colors.append('red')    
    else:
        colors.append('C0')  

fig, ax = plt.subplots(figsize = (16,8))

ax.bar(df_livros_geral['mes_str'], df_livros_geral['quantidade'], color=colors)

plt.xticks(rotation=70, ha='right')


ax.set_xlabel('Ano/Mês', labelpad = 10)
ax.set_ylabel('Quantidade')
ax.set_title('Evolução mensal - Livros em geral')
plt.grid(False)
plt.tight_layout()

plt.savefig('livros_geral.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
maior = df_livros_geral[df_livros_geral['quantidade'] == df_livros_geral['quantidade'].max()][['mes_str', 'quantidade']]
maior_mes = maior.iloc[0]['mes_str']
maior_qtd = maior.iloc[0]['quantidade']

menor = df_livros_geral[df_livros_geral['quantidade'] == df_livros_geral['quantidade'].min()][['mes_str', 'quantidade']]
menor_mes = menor.iloc[0]['mes_str']
menor_qtd = menor.iloc[0]['quantidade']

print(f'O mês com maior busca foi {maior_mes}, com {maior_qtd} buscas.')
print(f'O mês com menor busca foi {menor_mes}, com {menor_qtd} buscas.')
```

```python
df_livros_geral['ano'] = df_livros_geral['mes'].dt.year

media_geral = df_livros_geral.groupby('ano')['quantidade'].mean()

print("Média anual de buscas - Livros em geral:")
print(media_geral.round(2))
print()
```

```python
import pandas as pd
df_todos = pd.concat([df_livros_geral, df_livros_romance, df_livros_autoajuda], ignore_index=True)
media_ano_tipo = df_todos.groupby(['ano', 'tipo'])['quantidade'].mean().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

tipos = media_ano_tipo['tipo'].unique()

for tipo in tipos:
    dados_tipo = media_ano_tipo[media_ano_tipo['tipo'] == tipo]
    plt.plot(dados_tipo['ano'], dados_tipo['quantidade'], marker='o', label=tipo)

plt.title('Média anual de quantidade por tipo')
plt.xlabel('Ano')
plt.ylabel('Média da quantidade')
plt.legend(title='Tipo')

plt.savefig('media_anual_comparativo.png', dpi=300)

plt.show()
```
